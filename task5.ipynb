```python
#  Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("C:/Users/Ravali/Downloads/train.csv")
```

```python
#  Basic Information
print(df.head())
print(df.info())
print(df.describe())
```

```python
# ✏️ Observation:
# - Age has missing values.
# - Cabin has many missing values.
# - Most features are numeric or categorical.
```

```python
#  Missing Values
print(df.isnull().sum())
```

```python
# ✏️ Observation:
# - Around 177 missing values in Age.
# - Around 687 missing values in Cabin.
```

```python
# Value counts for some columns
print(df['Survived'].value_counts())
print(df['Pclass'].value_counts())
print(df['Sex'].value_counts())
```

```python
# Visual Exploration
```

```python
# Age Distribution
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# ✏️ Observation:
# - Most passengers are between 20 and 40 years old
```

```python
# Survival Count
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
# ✏️ Observation:
# - More people did not survive than survived.
```

```python
# Passenger Class vs Survival
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Passenger Class vs Survival')
plt.show()
```

```python
# ✏️ Observation:
# - Passengers from 1st class survived more than others.
```

```python
# Sex vs Survival
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Sex vs Survival')
plt.show()
```

```python
# ✏️ Observation:
# - Females had a much higher survival rate than males.
```

```python
# Boxplot: Age vs Pclass
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age vs Passenger Class')
plt.show()
```

```python
# ✏️ Observation:
# - Median age for 1st class is higher than 3rd class.
```

```python
# Scatterplot: Fare vs Age
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Fare vs Age by Survival')
plt.show()
```

```python
# ✏️ Observation:
# - Higher fare passengers had better survival chances.
```

```python
# Heatmap: Correlation Matrix (Fixed)
plt.figure(figsize=(10,8))
numeric_df = df.select_dtypes(include=[np.number])  # Select only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# ✏️ Observation:
# - Fare and Survived are positively correlated.
# - Pclass and Fare are negatively correlated.
```

```python
# Pairplot
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']])
plt.show()
```

```python
# ✏️ Observation:
# - Visible clusters for survival across age and fare.
```

```python
# Final Summary
print("""
Summary of Findings:
- Females had much higher survival chances than males.
- 1st class passengers survived more often than 3rd class.
- Age and Fare are important factors for survival.
- Data has missing values, especially in 'Age' and 'Cabin'.
- Fare is positively correlated with survival.
""")
```
